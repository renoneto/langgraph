# Reflexion Agent

This is an upgrade of the Reflection Agent, we will use Search API and Function Calling + Prompt Techniques to create a more advanced agent.

Architecture of the Reflexion Agent:
<img src="../images/reflexion_agent.jpg" alt="Reflection Agent" width="800"/>

Steps:
1. Start with a Respoder that contains the Response, Critique and Search Queries to improve the response.
2. Then we use a tool to execute the Search Queries (Tavily)
3. The Revisor Node will take the initial response and the search responses and then improve the response + critique it as well + new search queries + citations
4. This goes back to the Tool to search again and get more information.
5. The process continues until the response is satisfactory or a certain number of iterations is reached.
6. Finally, the agent will output the final response with citations.

The goal is for the agent to create an article with quality. 

Files
`schemas.py`: This file contains the schemas for the Reflexion Agent, including the `Reflection`, `AnswerQuestion` and `ReviseAnswer` classes.
`tool_executor.py`: This file contains the code for the tool executor that will execute the search queries and return the results.
- Concepts:
  - Using `ToolNode` which is a class that allows the agent to execute tools (like search APIs) and get results. We define only one tool node for the Search API.
  - To run the searches in batch we need to create a list of messages containing the `tool_calls` with the name of the tool and arguments. Moreover, the "id" is what LangGraph uses to connect the tool node with the responder and revisor nodes.
  - We then parse the responses and construct a list of `ToolMessage` objects that contain the results of the search queries.
- Search API: An API that allows the agent to perform searches based on the queries generated by the responder and revisor nodes.
`chains.py`: This file contains the chain definitions for the Reflexion Agent, including the responder and revisor chains.
`main.py`: This is where we construct the graph and run it
